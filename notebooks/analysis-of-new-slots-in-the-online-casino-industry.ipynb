```python
data = pd.read_csv('new_slots_data_2026.csv')
data.head()
```

## Data Analysis
Let's analyze some key statistics, including the average RTP and the distribution of player counts across these new slots. Understanding these metrics can help identify which games are performing well in the market.
We will also explore the correlation between the average bet amount and RTP.

```python
average_rtp = data['RTP'].mean()
average_player_count = data['player_count'].mean()
plt.figure(figsize=(12,6))
plt.hist(data['RTP'], bins=20, color='blue', alpha=0.7, label='RTP Distribution')
plt.title('Distribution of RTP for New Slots in 2026')
plt.xlabel('RTP (%)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
