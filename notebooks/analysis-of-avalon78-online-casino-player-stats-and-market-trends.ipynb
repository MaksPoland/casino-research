# Analysis of Avalon78 Online Casino: Player Stats and Market Trends
This notebook presents a comprehensive analysis of player statistics and market trends for the online casino, Avalon78. We will delve into various metrics, including player performance, game return-to-player (RTP) rates, and overall market trends in the gambling industry.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset containing player stats and game RTP
data = pd.read_csv('avalon78_player_data.csv')
data.head()
```

## Understanding Player Statistics
In this section, we will analyze player statistics to understand player behavior and performance in the Avalon78 casino. We will look at metrics such as average bet size, total wins, and the number of games played.

```python
# Calculating average metrics per player
player_stats = data.groupby('player_id').agg({
    'bet_size': 'mean',
    'total_wins': 'sum',
    'games_played': 'count'
}).reset_index()
player_stats.head()
```

### Distribution of Average Bet Size
Next, we will visualize the distribution of the average bet size among players. This provides insight into how much players are willing to stake on average.

```python
plt.figure(figsize=(10, 6))
sns.histplot(player_stats['bet_size'], bins=30, kde=True)
plt.title('Distribution of Average Bet Size Among Players')
plt.xlabel('Average Bet Size')
plt.ylabel('Frequency')
plt.show()
```

## Game RTP Analysis
In this section, we will analyze the Return to Player (RTP) percentages for various games offered by Avalon78. Game RTP is a critical metric that affects player retention and overall satisfaction.
According to [Discover Avalon78: Your 2026 Guide to Big Wins](https://topslotstoday.com/blog/avalon78), understanding RTP can significantly enhance players' gaming experience.

```python
# Load the game RTP data
rtp_data = pd.read_csv('avalon78_rtp_data.csv')
rtp_data.head()
```

### Visualizing Game RTP
We will visualize the RTP percentages for each game to identify which games offer higher returns to players.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='game_name', y='rtp_percentage', data=rtp_data)
plt.title('RTP Percentages for Games at Avalon78')
plt.xlabel('Game Name')
plt.ylabel('RTP Percentage')
plt.xticks(rotation=45)
plt.show()
```
