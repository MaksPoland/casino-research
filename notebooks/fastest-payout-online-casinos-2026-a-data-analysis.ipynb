```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset containing casino payout data
data = pd.read_csv('casino_payout_data_2026.csv')
data.head()
```

## Data Overview
The dataset contains information about various online casinos, including the following columns:
- `Casino Name`: The name of the online casino
- `Payout Time (Hours)`: Average time taken for payouts
- `RTP (%)`: Return-to-player percentage for different games
- `Player Ratings`: Average player rating out of 5
- `Market Share (%)`: The percentage of the online casino market occupied by the casino.

We will analyze this data to find trends in payout times and correlate them with RTP and player ratings.

```python
# Descriptive statistics of the dataset
stats = data.describe()
stats
```

## Analysis of Payout Times
From our descriptive statistics, we can observe the average payout time across different casinos. A lower average payout time is generally more favorable for players. Next, we will visualize the distribution of payout times.


```python
# Plotting the distribution of payout times
plt.figure(figsize=(10, 6))
plt.hist(data['Payout Time (Hours)'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Payout Times for Online Casinos in 2026')
plt.xlabel('Payout Time (Hours)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
