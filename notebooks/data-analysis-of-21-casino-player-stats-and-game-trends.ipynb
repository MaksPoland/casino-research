# Data Analysis of 21 Casino: Player Stats and Game Trends
This notebook aims to analyze various aspects of the 21 Casino, with a focus on player statistics, game return-to-player (RTP) rates, and market trends. Our goal is to provide insights that can help improve player engagement and optimize game offerings.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset containing player stats and game data
data = pd.read_csv('21_casino_data.csv')
data.head()
```

## Overview of the Dataset
The dataset contains information about player statistics, including the total bets, wins, losses, and the RTP for various games. Understanding this data will help us analyze player behavior and the performance of different games offered by 21 Casino.

```python
# Data cleaning and preprocessing
data.dropna(inplace=True)
data['RTP'] = data['Total Wins'] / data['Total Bets']
data.describe()
```

## Player RTP Analysis
By calculating the RTP for each game, we can identify which games are performing well in terms of player returns. Games with higher RTP percentages are generally more appealing to players. This analysis is crucial for understanding current market trends and potential game improvements.

```python
# Visualizing the RTP of games
plt.figure(figsize=(12, 6))
plt.bar(data['Game Name'], data['RTP'], color='blue')
plt.xlabel('Game Name')
plt.ylabel('Return to Player (RTP)')
plt.title('RTP of Games at 21 Casino')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
