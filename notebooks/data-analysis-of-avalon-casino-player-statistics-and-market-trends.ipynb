```python
# Analyze betting trends
betting_trends = player_data.groupby('date')['total_bets'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.plot(betting_trends['date'], betting_trends['total_bets'])
plt.title('Total Bets Over Time at Avalon Casino')
plt.xlabel('Date')
plt.ylabel('Total Bets')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
