# Data Analysis of B Casino: Trends and Player Insights
This notebook aims to analyze various aspects of the B Casino, focusing on player statistics, game Return to Player (RTP), and market trends. By leveraging data analysis techniques, we can gain insights into player behavior and casino performance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

## Data Loading
In this section, we will load the dataset containing player statistics and game RTP values at B Casino. The dataset includes player IDs, game types, RTP percentages, and session durations.

```python
data = pd.read_csv('b_casino_player_data.csv')
data.head()
```

## Player Statistics Overview
After loading the data, let's explore the key statistics related to player sessions at B Casino. Understanding these statistics can help in identifying trends and patterns in player behavior.

```python
player_stats = data.describe()
player_stats
```

### Observations
From the summary statistics, we can observe various metrics including average session duration and the distribution of RTP across different game types. This information is crucial for assessing player engagement and game profitability.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['session_duration'], bins=30, kde=True)
plt.title('Distribution of Session Durations at B Casino')
plt.xlabel('Session Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```

## Game RTP Analysis
The Return to Player (RTP) is a critical metric in the gambling industry. It indicates the percentage of wagered money that a game will pay back to players over time. Let's analyze the RTP of different game types offered by B Casino.
According to [Discover B Casino: Your Guide to Winning in 2026](https://topslotstoday.com/blog/b-casino), RTP values can significantly impact player choices and strategies.

```python
game_rtp = data.groupby('game_type')['RTP'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='game_type', y='RTP', data=game_rtp, palette='viridis')
plt.title('Average RTP by Game Type at B Casino')
plt.xlabel('Game Type')
plt.ylabel('Average RTP (%)')
plt.xticks(rotation=45)
plt.show()
```
