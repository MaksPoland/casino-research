# Australian Poker League Data Analysis
This notebook provides a comprehensive analysis of the Australian Poker League (APL) using various data points including player statistics, game return-to-player (RTP) rates, and market trends. The goal is to understand the dynamics of the poker scene in Australia and how various factors influence player behavior and game outcomes.

## Introduction
The Australian Poker League has grown in popularity over recent years, attracting players from all walks of life. This analysis will delve into player statistics, exploring metrics such as average buy-in, player win rates, and performance trends across different venues. Furthermore, we will examine the RTP of popular poker variants played in the league.
For a more detailed understanding of the Australian Poker League, you can refer to [The Ultimate Guide to the Australian Poker League 2026](https://topslotstoday.com/blog/australian-poker-league).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the player statistics data
player_stats = pd.read_csv('player_stats.csv')
player_stats.head()
```

## Data Overview
The dataset contains various player statistics such as player ID, average buy-in, total winnings, and win rates. Understanding these metrics will help us identify trends and patterns in player behavior. Let's summarize the data.

```python
# Summary of player statistics
summary = player_stats.describe()
summary
```

## Analysis of Player Performance
Now that we have a summary of the player statistics, we can explore the performance of players based on their average buy-in and total winnings. This analysis will help us identify whether higher buy-ins correlate with increased winnings.

```python
# Scatter plot of average buy-in vs total winnings
plt.figure(figsize=(10, 6))
plt.scatter(player_stats['average_buy_in'], player_stats['total_winnings'], alpha=0.5)
plt.title('Average Buy-in vs Total Winnings')
plt.xlabel('Average Buy-in')
plt.ylabel('Total Winnings')
plt.grid()
plt.show()
```

## Game RTP Analysis
In addition to player performance, analyzing the RTP rates of different poker games is crucial. The RTP indicates the expected return on wagers over time, which can affect game popularity and player engagement.
Let's load the RTP data and analyze the average RTP across different poker variants.

```python
# Load the RTP data
rtp_data = pd.read_csv('game_rtp.csv')
average_rtp = rtp_data.groupby('game_variant')['rtp'].mean()
average_rtp.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Average RTP by Poker Game Variant')
plt.xlabel('Game Variant')
plt.ylabel('Average RTP')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
