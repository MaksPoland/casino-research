# Best Interac e-Transfer Casinos in Canada
This notebook presents an analysis of the best online casinos in Canada that accept Interac e-Transfer as a payment method. We will explore player statistics, game return-to-player (RTP) rates, and market trends to identify the top casinos for players in Canada.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset containing casino data
casino_data = pd.read_csv('canadian_casinos_data.csv')
casino_data.head()
```

## Overview of Casino Data
The dataset includes various attributes such as casino name, game offerings, player demographics, and RTP percentages. We will analyze these attributes to determine the best Interac e-Transfer casinos in Canada.

```python
# Data Cleaning
casino_data.dropna(inplace=True)  # Remove missing values
casino_data['RTP'] = casino_data['RTP'].astype(float)

# Summary statistics
summary_stats = casino_data.describe()
summary_stats
```

## Statistical Analysis
The summary statistics provide insights into the average RTP of games offered by the casinos. Higher RTP values typically indicate better returns for players. We will visualize the distribution of RTP across the best casinos.

```python
# Plotting the distribution of RTP
plt.figure(figsize=(10, 6))
plt.hist(casino_data['RTP'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Game RTP in Canadian Casinos')
plt.xlabel('Return to Player (RTP) %')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
